```python
# Perfrom binary classification on diabetes dataset by using Support Vector Classifier 
#Apply Support vector classifier with linear and polynomial kernel by using cross validation to find the best value for hyperparameter degree 
#Apply radial basics kernel use cross validation to find the best value for gamma
```

```python
import pandas as pd
from sklearn.svm import SVC   
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier

```

```python
df=pd.read_csv("pima-indians-diabetes.csv",header=None)
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25, random_state=1)
```

```python
model1=SVC(kernel='linear')
```

```python
model1.fit(xtrain,ytrain)
```

```python
ypred=model1.predict(xtest)
```

```python
accuracy=accuracy_score(ytest,ypred)
accuracy
```

```python
model2=SVC(kernel='poly')
model2
```

```python
model2.fit(xtrain,ytrain)
ypred2=model2.predict(xtest)
```

```python
accuracy_score(ypred2,ytest)
```

```python
from sklearn.model_selection import RepeatedKFold
cv=RepeatedKFold(n_splits=10,n_repeats=3,random_state=1)
```

```python
parms={'degree':[5,6,7,8,9]}
from sklearn.model_selection import GridSearchCV
search = GridSearchCV(model2,parms,cv=cv)
result=search.fit(x,y)
best_parms = result.best_params_
print(best_parms)
```

```python
pmodel2=SVC(kernel='poly',degree=7)
pmodel2
```

```python
pmodel2.fit(xtrain,ytrain)
pypred2=pmodel2.predict(xtest)
```

```python
accuracy_score(pypred2,ytest)
```

```python
model3=SVC(kernel='rbf')
model3
```

```python
model3.fit(xtrain,ytrain)
ypred3=model3.predict(xtest)
```

```python
accuracy_score(ypred3,ytest)
```

```python
parms={'gamma':[0.1,0.01,0.001,0.0001,0.0001]}
from sklearn.model_selection import GridSearchCV
search= GridSearchCV(model3,parms,cv=cv)
result=search.fit(x,y)
best_params = result.best_params_
print(best_params)
```

```python
model3=SVC(kernel='rbf',gamma=0.0001)
model3
model3.fit(xtrain,ytrain)
ypred3= model3.predict(xtest)
accuracy_score(ypred3,ytest)
```
